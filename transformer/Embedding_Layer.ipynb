```python
from transformers import GPT2Tokenizer

# Initialize the GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Your input sentence
sentence = "I love you all"

# Tokenize the sentence
tokens = tokenizer.encode(sentence, add_special_tokens=True)

# Print the token ids
print("Token IDs:", tokens)

# Decode the tokens back to the sentence
decoded_sentence = tokenizer.decode(tokens)
print("Decoded Sentence:", decoded_sentence)

print("vocab size " , tokenizer.vocab_size)
```

```python
import torch
from torch import nn

token_tensor = torch.tensor([tokens])
print(token_tensor)

embedding_dim = 20
embedding_layer = nn.Embedding(num_embeddings=tokenizer.vocab_size, embedding_dim=embedding_dim)

# Pass the tokens through the embedding layer
embedded_tokens = embedding_layer(token_tensor)

print("Embedded Tokens Shape:", embedded_tokens.shape)
```
